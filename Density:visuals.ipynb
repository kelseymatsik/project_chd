**Density Plots for the Test Data**

```python
! git clone https://github.com/kelseymatsik/project_chd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/project_chd/fhs_test.csv')
df.head()
```

```python
sns.kdeplot(df['TenYearCHD'])
plt.title('Ten Year CHD Risk Density Plot')
```

```python
sns.kdeplot(data = df, x = 'TenYearCHD', hue = 'currentSmoker')
plt.title('Ten Year CHD Risk Based on Current Smoking Habits')
plt.xlabel('Ten Year CHD Risk')
plt.ylabel('Density')
plt.legend(title='Smoker or Non-Smoker', labels=['Non-Smoker', 'Smoker'])
```

```python
sns.kdeplot(data = df, x = 'TenYearCHD', hue = 'cigsPerDay')
plt.title('Ten Year Risk of CHD across Cigarettes Per Day ')
plt.xlabel('Ten Year CHD Risk')
plt.ylabel('Cigarettes Per Day')
```

```python
sns.kdeplot(data = df, x = 'TenYearCHD', hue = 'sex')
plt.title('Ten Year Risk of CHD across Sex ')
plt.xlabel('Ten Year CHD Risk')
plt.ylabel('sex')
plt.legend(title = 'Sex', labels = ['Female', 'Male'])
```

```python
#grouped kde of cigs per day and sex
df['#dailyCigsAndSex'] = df['cigsPerDay'] +df['sex']
sns.kdeplot(data=df, x='TenYearCHD', hue = '#dailyCigsAndSex')
```

```python
sns.kdeplot(data = df, x = 'TenYearCHD', hue = 'education')
plt.title('Ten Year Risk of CHD Across Education levels ')
plt.xlabel('Ten Year CHD Risk')
plt.ylabel('Education')
plt.legend(title = 'Education Level', labels = ['High School', 'High School/GED', 'Some College/Vocational', 'College'])
```

```python
sns.kdeplot(data = df, x = 'TenYearCHD', hue = 'prevalentHyp')
plt.title('Ten Year Risk of CHD across Hypertension levels ')
plt.xlabel('Ten Year CHD Risk')
plt.ylabel('Hypertension')
plt.legend (title = 'Defined as Hypertensive', labels = ['Yes', 'No'])
```

```python
#age
sns.kdeplot(data = df, x = 'TenYearCHD', hue = 'age')
plt.title('Ten Year Risk of CHD Across Age ')
plt.xlabel('Ten Year CHD Risk')
plt.ylabel('Age')
```

```python
#glucose
sns.kdeplot(data = df, x = 'TenYearCHD', hue = 'glucose')
plt.title('Ten Year Risk of CHD Across Glucose Levels ')
plt.xlabel('Ten Year CHD Risk')
plt.ylabel('Glucose')
```

```python
#cholesterol
sns.kdeplot(data = df, x = 'TenYearCHD', hue = 'totChol')
plt.title('Ten Year Risk of CHD Across Total Cholesterol (mg/dL) ')
plt.xlabel('Ten Year CHD Risk')
plt.ylabel('Total Cholesterol (mg/dL)')
```

```python
#groupedkde of cholesterol and hypertension
df['chlandHyp'] = df['age'] +df['education']
sns.kdeplot(data=df, x='prevalentHyp', hue = 'totChol')
plt.title
```

```python
#grouped kde of education and hypertesnion -> need to figure out labels
df['educationAndHyp'] = df['education'] +df['prevalentHyp']
sns.kdeplot(data=df, x='TenYearCHD', hue = 'educationAndHyp')
plt.legend(title='Education and Hypertension', labels=[''])
```
