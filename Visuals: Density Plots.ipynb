**Density Plots **

```python
! git clone https://github.com/kelseymatsik/project_chd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/project_chd/fhs_test.csv')
df.head()
```

```python
#cigs per day and risk
sns.kdeplot(data = df, x = 'cigsPerDay', hue = 'TenYearCHD')
plt.title('Ten Year CHD Risk Based on Current Smoking Habits')
plt.xlabel('Cigarettes Per Day')
plt.ylabel('Density')
plt.legend(title='CHD Diagnosis', labels=['No', 'Yes'])
```

```python
#current smoking habits and risk
sns.kdeplot(data = df, x = 'TenYearCHD', hue = 'currentSmoker')
plt.title('Ten Year CHD Risk Based on Current Smoking Habits')
plt.xlabel('Ten Year CHD Risk')
plt.ylabel('Density')
plt.legend(title='Smoker or Non-Smoker', labels=['Non-Smoker', 'Smoker'])
```

```python
# cigs smoked per day and sex
sns.kdeplot(data = df, x = 'cigsPerDay', hue = 'sex')
plt.title('cigsPerDay across Sex ')
plt.xlabel('Cigarettes Per Day')
plt.ylabel('Density')
plt.legend(title='Sex', labels=['Female', 'Male'])

```

```python
#sex and risk
sns.kdeplot(data = df, x = 'TenYearCHD', hue = 'sex')
plt.title('Ten Year Risk of CHD across Sex ')
plt.xlabel('Ten Year CHD Risk')
plt.ylabel('sex')
plt.legend(title = 'Sex', labels = ['Female', 'Male'])
```

```python
#education and risk
sns.kdeplot(data = df, x = 'education', hue = 'TenYearCHD')
plt.title('Ten Year Risk of CHD Across Education levels ')
plt.xlabel('Education')
plt.ylabel('Density')
plt.legend(title = 'CHD Diagnosis', labels = ['No', 'Yes'])
```

```python
#hypertension and risk
sns.kdeplot(data = df, x = 'TenYearCHD', hue = 'prevalentHyp')
plt.title('Ten Year Risk of CHD across Hypertension levels ')
plt.xlabel('Ten Year CHD Risk')
plt.ylabel('Hypertension')
plt.legend (title = 'Defined as Hypertensive', labels = ['No', 'Yes'])
```

```python
#age and risk
sns.kdeplot(data = df, x = 'age', hue = 'TenYearCHD')
plt.title('Ten Year Risk of CHD Across Age ')
plt.xlabel('Age at Time of Exam')
plt.ylabel('Density')
plt.legend(title = 'CHD Diagnosis', labels  = ['No', 'Yes'])
```

```python
#glucose and risk
sns.kdeplot(data = df, x = 'glucose', hue = 'TenYearCHD')
plt.title('Glucose Levels and Ten Year CHD Risk  ')
plt.xlabel('Glucose Levels')
plt.ylabel('Density')
plt.legend(title = 'CHD Diagnosis', labels  = ['No', 'Yes'])
```

```python
#cholesterol and risk
sns.kdeplot(data = df, x = 'totChol', hue = 'TenYearCHD')
plt.title('Ten Year Risk of CHD Across Total Cholesterol (mg/dL) ')
plt.xlabel('Total Cholesterol (mg/dL)')
plt.ylabel('Density')
plt.legend(title = 'CHD Diagnosis', labels  = ['No', 'Yes'])
```

```python
#kde of cholesterol and hypertension
sns.kdeplot(data = df, x = 'totChol', hue = 'prevalentHyp')
plt.title('Cholesterol and Hypertension Levels ')
plt.xlabel('Total Cholesterol (mg/dL)')
plt.ylabel('Density')
plt.legend(title = 'Defined as Hypertensive and Treated', labels  = ['No', 'Yes'])
```

```python
#education and hypertension
sns.kdeplot(data = df, x = 'education', hue = 'prevalentHyp')
plt.title('Education and Hypertension Levels ')
plt.xlabel('Education Levels')
plt.ylabel('Density')
plt.legend(title = 'Hypertensive ', labels = ['No', 'Yes'])
```
